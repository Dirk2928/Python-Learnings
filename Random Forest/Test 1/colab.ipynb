```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis_encoded'] = le.fit_transform(df['diagnosis'])
print(le.classes_)  # ['B', 'M']
```

```python
# if diagnosis is in column 1 and is string:
numeric_df = df.drop(columns=['diagnosis'])  
corr = numeric_df.iloc[:, 0:10].corr()     
print(corr)

plt.figure(figsize=(10,8))
sb.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation (numeric columns only)")
plt.show()
```
